```python
import pandas as pd
import random
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('cointegrated/rubert-tiny2')
```

```python
df_cv = pd.read_csv("../data/datasets_with_text/cv.csv")
df_vac = pd.read_csv("../data/datasets_with_text/vac.csv")
```

```python
cv_num = random.randint(0, len(df_cv) - 1)
cv = df_cv.iloc[cv_num]["cv_text"]
```

```python
print(cv)
```

```python
kw_model = KeyBERT(model=model)
```

```python
kw_model.extract_keywords(cv, keyphrase_ngram_range=(1, 3), use_mmr=True, diversity=0.9, top_n=10)
```
