```python
import pandas as pd

from navec import Navec
from slovnet import NER
from razdel import tokenize

import pymorphy3
```

Попробуем достать локации из текста с помощью natasha

```python
morph = pymorphy3.MorphAnalyzer()
```

```python
df_cv = pd.read_csv("../data/datasets_with_text/cv.csv")

df_vac = pd.read_csv("../data/datasets_with_text/vac.csv")
```

```python
print(df_vac.iloc[0]["vac_text"])
```

```python
navec = Navec.load('./models/navec_news_v1_1B_250K_300d_100q.tar')
ner = NER.load('./models/slovnet_ner_news_v1.tar')
ner.navec(navec)
```

```python
df_cv["locations"] = df_cv["cv_text"].apply(lambda x: [loc for loc in set([morph.parse(x[span.start:span.stop])[0].normal_form for span in ner(x).spans if span.type=="LOC"])])
df_vac["locations"] = df_vac["vac_text"].apply(lambda x: [loc for loc in set([morph.parse(x[span.start:span.stop])[0].normal_form for span in ner(x).spans if span.type=="LOC"])])
```

```python
df_cv.to_csv("../data/datasets_with_locations/cv.csv")
df_vac.to_csv("../data/datasets_with_locations/vac.csv")
```
